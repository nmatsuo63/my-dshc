## 元画像を表示させる方法
元画像は、28ピクセル*28ピクセルのサイズで、0から255までのグレースケール画像です。

```python
for i in range(15):
    count = 0
    for ind in range(1000):
        # カタカナ番号
        num = np.where(train_label[ind])[0][0]
        if num==i:
            img = train_data[ind]
            label = train_label[ind]    
            print(label)
            plt.imshow(img[0,:,:], cmap='gray')
            plt.show()
            count += 1
            if count==2:
                break
```
