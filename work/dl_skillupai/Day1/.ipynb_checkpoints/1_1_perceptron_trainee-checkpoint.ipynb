# パーセプトロン

### ANDゲート(シンプルな実装方法)

```python
def AND(x1, x2):
    w1, w2, theta = 0.5,0.5,0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    else:
        return 1

print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

### ANDゲーとの実装(NNでの利用を意識した実装方法)
閾値$\theta$をバイアス項$b$として扱っている

```python
import numpy as np

def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    
    tmp = np.dot(x, w) + b
    if tmp <= 0:
        return 0
    else:
        return 1

print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

### XORゲート

### [演習]
* XORゲートを完成させましょう。\***の場所をAND関数またはOR関数またはNAND関数に置き換えてください。

```python
def XOR(x1, x2):
    s1 = ***(x1,x2)
    s2 = ***(x1,x2)
    y = ***(s1, s2)
    return y

print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
        
```
